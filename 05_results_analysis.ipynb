```python
def plot_metric(metrics, column, type_):
    metric_= metrics['VAR(O,S)'][column].to_frame()
    metric_.columns= ['VAR(O,S)']
    metric_['ARIMA(O)']= metrics['ARIMA(O)'][column]
#    metric_['VARMAX']= metrics['VARMAX'][column]
    metric_['LSTM(OS)']= metrics['LSTM(OS)'][column]
    metric_['LSTM(O)']= metrics['LSTM(O)'][column]
    
    ax=metric_.drop(index='global').plot(grid=True, figsize=(10,5), kind='bar');
    ax.set_xlabel('Time horizon (m)', fontsize=14)
    ax.set_ylabel(column, fontsize=14)
    ax.set_xticks(range(0,8),'15 30 45 60 75 90 105 120'.split())
    
    plt.tight_layout()
    plt.savefig(os.path.join('figs',f'{type_}_{column}.png'))
```

```python
plot_metric(metrics, 'MAPE', 'city_level')
```

```python
plot_metric(metrics, 'CVRMSE', 'city_level')
```

```python
plot_metric(metrics, 'MSE', 'city_level')
```

```python
plot_metric(metrics, 'RMSE', 'city_level')
```

```python
plot_metric(metrics, 'MAE', 'city_level')
```
